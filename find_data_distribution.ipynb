```python
import seaborn as sns
import pandas as pd
from fitter import Fitter,get_common_distributions, get_distributions
```

```python
df = pd.read_csv('data.csv')
```

```python
df.columns
```

```python
def find_distribution(data: pd.Series):
    f = Fitter(data, distributions=get_common_distributions())
    f.fit(max_workers=1)
    f.summary()
```

```python
find_distribution(df['Delivery_Time'])
```

```python
df.columns
find_distribution(df['Customer_Rating'])
```
